# Context
## Problem Statement

**Customer Personality Analysis is a detailed analysis of a company’s ideal customers. It helps a business to better understand its customers and makes it easier for them to modify products according to the specific needs, behaviors and concerns of different types of customers.**

**Customer personality analysis helps a business to modify its product based on its target customers from different types of customer segments. For example, instead of spending money to market a new product to every customer in the company’s database, a company can analyze which customer segment is most likely to buy the product and then market the product only on that particular segment**.

```
from scipy.stats import kruskal
Graduation_Edu = Hypothes_Df[Hypothes_Df['Education']=='Graduation']['Income']
Master_Edu = Hypothes_Df[Hypothes_Df['Education']=='Master']['Income']
Phd_Edu = Hypothes_Df[Hypothes_Df['Education']=='PhD']['Income']
```

```
Master_Edu.plot.density()
plt.show()
#Master Education data distribution can be normal distribution 
#then we do parametric test
```

### - **Fail to Reject H0: Sample distributions are equal.**
### - **Reject H0: Sample distributions are not equal.**

```
from scipy.stats import ttest_1samp
Stats, P = ttest_1samp(Master_Edu,(Master_Edu.mean()+50))
a = 0.05
print(f'Statistics = {Stats:.2f}\nP-Value = {P:.2f}')
if P > a:
    print("H0 Fail to  Reject  Mean <<(Same Distributions)>> Almost Equal")
else:
    print('H0 Reject  Mean <<(Different Distributions)>> Different')
```

```
from scipy.stats import mannwhitneyu

Stats, P = mannwhitneyu(Phd_Edu, Graduation_Edu)
a = 0.05
print(f'Statistics = {Stats:.2f}\nP-Value = {P:.2f}')
if P > a:
    print("H0 Fail to  Reject All Median <<(Same Distributions)>> are Almost Equal")
else:
    print('H0 Reject All Median <<(Different Distributions)>> are Different')
```

#### - **H0:** **All Sample  Distrubiton Are the Same**
#### - **H1:** **One or More Distrubition are Different From Each Other**

```
Stats, P = kruskal(Graduation_Edu, Master_Edu, Phd_Edu)
print(f'Statistics = {Stats:.2f}\nP-Value = {P:.2f}')
a = 0.05
if P > a:
    print("H0 Fail to  Reject All Median <<(Same Distributions)>> are Almost Equal")
else:
    print('H0 Reject All Median <<(Different Distributions)>> are Different')
```
