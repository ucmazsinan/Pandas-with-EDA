```
from pprint import pprint

agg_dict = {field: ['mean', 'median'] for field in data.columns if field != 'species'}
agg_dict['petal_length'] = 'max'
pprint(agg_dict)
data.groupby('species').agg(agg_dict)
```

```
from pprint import pprint

agg_dict = {field:['mean','var', 'std'] for field in iris.columns if field != 'species'}
pprint(agg_dict)
round(iris.groupby('species').agg(agg_dict),2)
```

```
scatter_plot = plt.figure()
ax = scatter_plot.add_subplot(1,1,1)
ax.scatter(iris.sepal_length,
                     iris.sepal_width)
ax.set(xlabel = 'Sepal Length (cm)',
       ylabel = 'Sepal Width(cm)',
       title = 'Sepal Length and Sepal Width')
plt.show()
```

```
histogram_plot = plt.figure()
ax1 = histogram_plot.add_subplot(1,1,1)
ax1.hist(iris['petal_length'],bins=20)
ax1.set(xlabel='Petal Length (cm)', 
       ylabel='Frequency',
       title='Distribution of Petal Lengths')
plt.show()
```

```
sns.set_context('notebooks')
ax = iris.plot.hist(bins=25, alpha = 0.5)
ax.set_xlabel('Size (cm)')
```

```
axList = iris.hist(bins=25)
for ax in axList.flatten():
    if ax.is_last_row():
        ax.set_xlabel('Size (cm)')
    if ax.is_first_col():
        ax.set_ylabel('Frequency')
```

```
iris.boxplot(by = 'species')
```

```
plot_iris = (iris
             .set_index('species')
             .stack()
             .to_frame()
             .reset_index()
             .rename(columns={0:'size', 'level_1':'measurement'})
            )

plot_iris.head()
```

```
sns.set_style('white')
sns.set_context('notebook')
sns.set_palette('dark')

f = plt.figure(figsize=(6,4))
sns.boxplot(x='measurement', y='size', 
            hue='species', data=plot_iris)
plt.show()
```
